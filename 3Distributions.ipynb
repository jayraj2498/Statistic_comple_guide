# Distributions 

## Normal distribution

#### Normal distribution also known as Gaussian distribution, is a symmetric bell-shaped probability distribution. It is characterized by its mean (center) and standard deviation (spread). Many natural phenomena follow a normal distribution

Normal Distribution (Gaussian Distribution):

Bell-shaped curve.

Symmetric around the mean.

Many natural phenomena follow a normal distribution.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate and plot a normal distribution with smoothing (KDE)
mean = 0
std_dev = 1
data = np.random.normal(mean, std_dev, 1000)

# Create a histogram with smoothing (KDE)
sns.histplot(data, kde=True, bins=30, color='skyblue')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Or Guassian  (Kernel Density Estimation)')
plt.show()
```

## Standard Normal Distribution

### Standard Normal Distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It is a special case of the normal distribution and is commonly denoted as 
Z∼N(0,1).

Bell-shaped curve.

Mean  is 0.

Standard Deviation  is 1.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate and plot a standard normal distribution
data_standard_normal = np.random.standard_normal(1000)
plt.hist(data_standard_normal, bins=30, density=True, color='skyblue')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution')
plt.show()
```

## Z-score

### Z-score measures how many standard deviations a data point is far away  from the mean. It standardizes data, making it comparable across different scales. 


Z-score indicates position relative to mean.

Positive Z: Above mean.

Negative Z: Below mean.


```python
import numpy as np
from scipy.stats import zscore

# Generate data
data = np.random.normal(0, 1,10)

# Calculate Z-scores
z_scores = zscore(data) 

print(z_scores )
```

## Bernoulli distribution

Bernoulli distribution models binary outcomes (e.g., success/failure) with a single trial. It has a probability 

p of success and 

1−p of failure


Binary outcomes.

Probability p of success.

Visualized with a histogram or bar plot

```python
import numpy as np
from scipy.stats import bernoulli

# Generate Bernoulli data
p = 0.3  # Probability of success
bernoulli_data = bernoulli.rvs(p, size=1000)
```

```python
set(bernoulli_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Generate Bernoulli data
p = 0.3  # Probability of success
bernoulli_data = bernoulli.rvs(p, size=1000)

# Create a histogram or bar plot
plt.hist(bernoulli_data, bins=[-0.5, 0.5, 1.5], density=True, color='skyblue', edgecolor='black')
plt.xticks([0, 1], ['Failure', 'Success'])
plt.xlabel('Outcomes')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution')
plt.show()
```

## Binomial distribution


### Binomial distribution models the number of successes in a fixed number of independent and identical trials, each with a probability

### p of success.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Generate Binomial data
n = 5  # Number of trials
p = 0.3  # Probability of success
binomial_data = binom.rvs(n, p, size=1000)
```

```python
set(binomial_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Generate Binomial data
n = 5  # Number of trials
p = 0.3  # Probability of success
binomial_data = binom.rvs(n, p, size=1000)

# Create a histogram or bar plot
plt.hist(binomial_data, bins=np.arange(-0.5, n+1.5, 1), density=True, color='skyblue', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=5, p=0.3)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data for a normal distribution
data_normal = np.random.normal(0, 1, 1000)

# Plot histogram and fit a normal distribution
plt.hist(data_normal, bins=30, density=True, color='skyblue', edgecolor='black')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2)
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Generate data for a Poisson distribution
lambda_ = 3  # Average number of events
data_poisson = poisson.rvs(lambda_, size=1000)

# Plot histogram
plt.hist(data_poisson, bins=np.arange(-0.5, max(data_poisson) + 1.5, 1), density=True, color='skyblue', edgecolor='black')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.show()
```

```python
# 10. F-Distribution:
# Characteristics:

# Definition:

# The F-distribution is a probability distribution that arises in statistical hypothesis testing, particularly
# in the context of analysis of variance (ANOVA) and regression analysis.

# Used to compare variances between two or more groups.
# Essential in assessing the significance of the difference in variances.


# Used in analysis of variance (ANOVA) and regression analysis.
# Ratio of two chi-square distributions.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Generate data for an F-distribution
data_f = f.rvs(3, 5, size=1000)  # Example with degrees of freedom (d1=3, d2=5)

# Plot histogram
plt.hist(data_f, bins=30, density=True, color='skyblue', edgecolor='black')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = f.pdf(x, 3, 5)
plt.plot(x, p, 'k', linewidth=2)
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('F-Distribution')
plt.show()
```

```python
# Pareto Distribution: 
    
# The Pareto distribution is a power-law probability distribution often used to describe phenomena 
# where a small number of items have significantly higher occurrences than the majority. 

# Named after economist Vilfredo Pareto, who noted that 80% of Italy's wealth was owned by 20% of the population.

# Commonly used in fields like economics, physics, and social sciences to model wealth distribution
# , frequency of events,
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

# Generate data for a Pareto distribution
data_pareto = pareto.rvs(2.5, size=1000)  # Example with shape parameter (α) 2.5

# Plot histogram
plt.hist(data_pareto, bins=30, density=True, color='skyblue', edgecolor='black')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = pareto.pdf(x, 2.5)
plt.plot(x, p, 'k', linewidth=2)
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Pareto Distribution')
plt.show()
```
